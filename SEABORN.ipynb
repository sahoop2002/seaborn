# LINE PLOT

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

var = [1,2,3,4,5,6,7,8,9]
var_1 = [3,6,7,8,3,2,1,7,8]
x_1 = pd.DataFrame({"var":var, "var_1":var_1})
sns.lineplot(x="var", y="var_1", data=x_1)
plt.show()
```

```python
var = [1,2,3,4,5,6,7,8,9]
var_1 = [3,6,7,8,3,2,1,7,8]

sns.lineplot(x=var, y=var_1)

plt.show()
```

```python
y_1 = sns.load_dataset("penguins")
y_1
```

```python
sns.lineplot(x="bill_length_mm", y="bill_depth_mm", data=y_1)
plt.show()
```

```python
# hue parameter

sns.lineplot(x="bill_length_mm", y="bill_depth_mm", data=y_1, hue ="sex")
plt.show()
```

```python
# figure size
plt.figure(figsize=(10,5))
sns.lineplot(x="bill_length_mm", y="bill_depth_mm", data=y_1, hue ="sex")
plt.show()
```

```python
# style parameter

sns.lineplot(x="bill_length_mm", y="bill_depth_mm", data=y_1, hue ="sex",style="sex")
plt.show()
```

```python
# palette parameter  >> if you want to know more about this you can refer this https://www.practicalpythonfordatascience.com/ap_seaborn_palette

sns.lineplot(x="bill_length_mm", y="bill_depth_mm", data=y_1, hue ="sex",style="sex",palette = "Accent")
plt.show()
```

```python
# markers parameter
plt.figure(figsize=(20,8))
sns.lineplot(x="bill_length_mm", y="bill_depth_mm", data=y_1, hue ="sex",style="sex",palette = "cool",markers=["o",">"])
plt.show()
```

```python
# dashes parameter [remove dotted line]
plt.figure(figsize=(20,8))
sns.lineplot(x="bill_length_mm", y="bill_depth_mm", data=y_1, hue ="sex",style="sex",palette = "cool",markers=["o",">"],dashes = False)
plt.show()
```

```python
# legend parameter [in matplotlib we can call with legend and in seaborn we can remove with legend ]
plt.figure(figsize=(20,8))
sns.lineplot(x="bill_length_mm", y="bill_depth_mm", data=y_1, hue ="sex",style="sex",
             palette = "cool",markers=["o",">"],dashes = False,legend = False)

plt.grid()
plt.title("Line Plot")
plt.show()
```

# BAR PLOT

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd


var = sns.load_dataset("penguins")
var
```

```python
sns.barplot(x="island", y="bill_length_mm", data=var)
plt.show()
```

```python
# another methode of plotting  :
sns.barplot(x=var.island, y=var.bill_length_mm)
plt.show()
```

```python
order_1 = ["Torgersen","Dream","Biscoe"]
sns.barplot(x="island", y="bill_length_mm", data=var, hue="sex",order=order_1)
plt.legend(loc='upper left')
plt.show()
```

```python
order_1 = ["Torgersen","Dream","Biscoe"]
sns.barplot(x="island", y="bill_length_mm", data=var, hue="sex",order=order_1,hue_order=["Female","Male"],palette="viridis")
plt.legend(loc='upper left')
plt.show()
```

```python
# errorbar[ci paramerter]
order_1 = ["Torgersen","Dream","Biscoe"]
sns.barplot(x="island", y="bill_length_mm", data=var, hue="sex",order=order_1,hue_order=["Female","Male"],palette="viridis",errorbar=('ci', 10))
plt.legend(loc='upper left')
plt.show()
```

```python
# n_boot parameter
order_1 = ["Torgersen","Dream","Biscoe"]
sns.barplot(x="island", y="bill_length_mm", data=var, hue="sex",order=order_1,hue_order=["Female","Male"],palette="viridis",n_boot=10)
plt.legend(loc='upper left')
plt.show()
```

```python
# orient parameter [to use orient parameter we have to take only numerical values]
plt.figure(figsize=(20,10))
order_1 = ["Torgersen","Dream","Biscoe"]
sns.barplot(x="bill_length_mm", y="flipper_length_mm",data=var,orient = "h")

plt.show()
```

```python
# color parameter, saturation parameter, errcolor and errwidth parameter[err_kws={'color': 'colorname','linewidth':10}]
sns.barplot(x="island", y="bill_length_mm", data=var,color= "blue",saturation = 0.3, err_kws={'color': 'red', 'linewidth': 10})
plt.show()
```

```python
# capsize parameter
sns.barplot(x="island", y="bill_length_mm", data=var,color= "yellow",err_kws={'color': 'blue'},capsize=1)
plt.show()
```

```python
# imposing 2 bar graph together [dodge parameter]
order_1 = ["Torgersen","Dream","Biscoe"]
sns.barplot(x="island", y="bill_length_mm", data=var, hue="sex",order=order_1,hue_order=["Female","Male"],palette="viridis",dodge=False)
plt.legend(loc='upper left')
plt.show()
```

```python
# alpha parameter [to fade the original color]
order_1 = ["Torgersen","Dream","Biscoe"]
sns.barplot(x="island", y="bill_length_mm", data=var, hue="sex",
            order=order_1,hue_order=["Female","Male"],palette="viridis",saturation=1,alpha=0.5)
plt.legend(loc='upper left')
plt.show()
```

```python
# grid in seaborn
sns.set_style("darkgrid")
# sns.set(style = "darkgrid")
order_1 = ["Torgersen","Dream","Biscoe"]
sns.barplot(x="island", y="bill_length_mm", data=var, hue="sex",
            order=order_1,hue_order=["Female","Male"],palette="viridis",saturation=1,alpha=0.5)
plt.legend(loc='upper left')
plt.show()
```

# HISTOGRAM PLOT

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
hist = sns.load_dataset("penguins")
hist
```

```python
# HISTOGRAM PLOT
sns.displot(hist.bill_length_mm)
plt.show()
```

```python
# bins parameter [to organize the data]
sns.displot(hist.bill_length_mm,bins=[30,35,40,45,50,55,60])
plt.show()
```

```python
# kde parameter [kernel density estimation]
sns.displot(hist.flipper_length_mm,bins=[170,180,190,200,210,220,230],color="b",kde = True)
plt.show()
```

```python
# rug parameter [A rug plot shows the individual data points as small vertical or horizontal ticks along the axis,
#                 providing a visual representation of the data's density.]
sns.displot(hist.flipper_length_mm,bins=[170,180,190,200,210,220,230],color="b",kde = True,rug=True)
plt.show()
```

```python
# log_scale parameter [To apply a logarithmic scale to an axis in Seaborn]
sns.displot(hist.flipper_length_mm,kde=True,log_scale=True)
plt.show()
```

# HEATMAP PLOT IN SEABORN
[HEATMAP WORKS ON 2 DIMENTION DATA]

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
X = np.linspace(1,10,100).reshape(10,10)
X
```

```python
sns.heatmap(X)
plt.show()
```

```python
heat = sns.load_dataset("car_crashes")
heat
```

```python
# drop function
heat = sns.load_dataset("car_crashes")
X = heat.drop(columns=["abbrev","ins_premium","ins_losses"],axis = 1)
X
```

```python
sns.heatmap(X)
plt.show()
```

```python
# vmax & vmin parameter
sns.heatmap(X,vmin=0,vmax=15)
plt.show()
```

```python
# cmap & annot parameter
figure = plt.figure(figsize=(20,10))
sns.heatmap(X,vmin=0,vmax=15,cmap="PuOr",annot = True)
plt.show()
```

```python
# how to put my own value in heatmap
X = np.linspace(1,10,10).reshape(2,5)
X
```

```python
sns.heatmap(X,vmin=0,vmax=15,cmap="PuOr",annot = True)
plt.show()
```

```python
ar = np.array([["a0","a1","a2","a3","a4"],["b0","b1","b2","b3","b4"]])
ar
```

```python
sns.heatmap(X,vmin=0,vmax=15,cmap="PuOr",annot = ar ,fmt="s")
plt.show()
```

```python
y = {"fontsize":20,"color":"b"}
sns.heatmap(X,vmin=0,vmax=15,cmap="PuOr",annot = True,annot_kws=y)
plt.show()
```

```python
# linewidth parameter
y = {"fontsize":20,"color":"b"}
sns.heatmap(X,vmin=0,vmax=15,cmap="PuOr",annot = True,annot_kws=y,linewidth=10)
plt.show()
```

```python
# linecolor & cbar parameter
y = {"fontsize":20,"color":"b"}
sns.heatmap(X,vmin=0,vmax=15,cmap="PuOr",annot = True,annot_kws=y,linewidth=10,linecolor = "g",cbar=False)
plt.show()
```

```python
# xticklabels & yticklabels parameter [hide or show labels]
y = {"fontsize":20,"color":"b"}
sns.heatmap(X,vmin=0,vmax=15,cmap="PuOr",annot = True,annot_kws=y,linewidth=10,linecolor = "g",
            cbar=False, xticklabels=False, yticklabels=False)
plt.show()
```

```python
# create labels through seaborn
y = {"fontsize":20,"color":"b"}
ax = sns.heatmap(X,vmin=0,vmax=15,cmap="PuOr",annot = True,annot_kws=y,linewidth=10,linecolor = "g",
           cbar=False, xticklabels=False, yticklabels=False)
ax.set(xlabel="CAR",ylabel = "MOTORCYCLE")
#increase or decrease the font size of labels
sns.set(font_scale = 3)
plt.show()
```
