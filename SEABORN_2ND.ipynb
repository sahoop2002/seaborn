# COUNT PLOT

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
var = sns.load_dataset("tips")
var
```

```python
# PLOTTING COUNT PLOT
sns.countplot(x="sex", data=var,hue='smoker')
plt.show()
```

```python
# if we change only axis, we can convert it vertical to horizontal
sns.countplot(y="sex", data=var,hue='smoker')
plt.show()
```

```python
# using color palette & saturation
sns.countplot(x="sex", data=var,hue='smoker',palette = "bwr",saturation = 1)
plt.show()
```

# VIOLIN PLOT

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
var1 = sns.load_dataset("tips")
var1
```

```python
sns.violinplot(x="day",y="total_bill",data=var1)
plt.show()
```

```python
sns.violinplot(x="day",y="total_bill",data=var1, hue="time")
plt.show()
```

```python
# linewidth parameter
sns.violinplot(x="day",y="total_bill",data=var1, hue="time", linewidth = 3)
plt.show()
```

```python
# color parameter
sns.violinplot(x="day",y="total_bill",data=var1, hue="time", linewidth = 2, palette = "viridis")
plt.show()
```

```python
# differentiate dinner and lunch
sns.violinplot(x="time",y="total_bill",data=var1, order = ["Dinner","Lunch"])
plt.show()
```

```python
# saturation parameter [control the color of darknessor lightness] & color parameter
sns.violinplot(x="day",y="total_bill",data=var1, saturation = 1, color = 'r')
plt.show()
```

```python
# split parameter with male and female
sns.violinplot(x="day",y="total_bill",data=var1, hue = "sex", split = True)
plt.show()
```

```python
# density_norm parameter ["count","width", and "area" we can put in density_norm parameter]
sns.violinplot(x="day",y="total_bill",data=var1, hue = "sex", split = True,density_norm = "width")
plt.show()
```

```python
# change it verticle to horizontal [just change the x and and y axis]
sns.violinplot(y="day",x="total_bill",data=var1, hue = "sex")
plt.show()
```

```python
# inner parameter ["quart","box","quartile","stick","point",None]
sns.violinplot(x="time",y="total_bill",data=var1, order=["Dinner","Lunch"],inner = "quart")
plt.show()
```

```python
# single violin plot [we can plot it with numerical value only]
sns.violinplot(y = "total_bill",data=var1)
plt.show()
```

# PAIR PLOT IN SEABORN

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
var2 = sns.load_dataset("tips")
var2
```

```python
sns.pairplot(var2)
plt.show()
```

```python
# differentiate with hue parameter in pair plot
sns.pairplot(var2,hue="sex")
plt.show()
```

```python
# if you want to take some particular data from your dataset [when you pass the string it should be in numerical value]
sns.pairplot(var2,vars=["total_bill","size"],hue="sex")
plt.show()
```

```python
# set the hue_order
sns.pairplot(var2,vars=["total_bill","size"],hue="sex",hue_order=["Female","Male"])
plt.show()
```

```python
# palette parameter for set the color
sns.pairplot(var2,vars=["total_bill","size"],hue="sex",hue_order=["Female","Male"],palette="viridis")
plt.show()
```

```python
# set x & y axis
sns.pairplot(var2,hue="sex",hue_order=["Female","Male"],palette="viridis",y_vars=["total_bill","tip"])
plt.show()
```

```python
# kind parameter ["scatter","kde","hist","reg"]
sns.pairplot(var2,hue="sex",hue_order=["Female","Male"],palette="viridis",kind="kde")
plt.show()
```

```python
# kind parameter ["scatter","kde","hist","reg"] with diag_kind parameter
sns.pairplot(var2,hue="sex",hue_order=["Female","Male"],palette="viridis",kind="kde",diag_kind="hist")
plt.show()
```

```python
# marker parameter
sns.pairplot(var2,hue="sex",hue_order=["Female","Male"],markers=["*",">"])
plt.show()
```

# STRIP PLOT

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
var = sns.load_dataset("tips")
var
```

```python
# strip plot
sns.stripplot(x="day",y="total_bill",data=var)
plt.show()
```

```python
# differentiate with hue parameter and legend placement
sns.stripplot(x="day",y="total_bill",data=var,hue="sex")
plt.legend(loc="upper left")
plt.show()
```

```python
# palette parameter
plt.figure(figsize=(10, 5))
sns.stripplot(x="day",y="total_bill",data=var,hue="sex",palette="seismic")
plt.legend(loc="upper left")
plt.show()
```

```python
# jitter parameter
plt.figure(figsize=(10, 5))
sns.stripplot(x="day",y="total_bill",data=var,hue="sex",palette="seismic",jitter =0.3)
plt.legend(loc="upper left")
plt.show()
```

```python
# size parameter
plt.figure(figsize=(15, 5))
sns.stripplot(x="day",y="total_bill",data=var,hue="sex",palette="seismic",jitter =0.3,size = 10)
plt.legend(loc="upper left")
plt.show()
```

```python
# marker parameter
plt.figure(figsize=(15, 5))
sns.stripplot(x="day",y="total_bill",data=var,hue="sex",palette="seismic",jitter =0.3,size = 10,marker="*")
plt.legend(loc="upper left")
plt.show()
```

```python
# alpha parameter [highlight the color]
plt.figure(figsize=(15, 5))
sns.stripplot(x="day",y="total_bill",data=var,hue="sex",palette="seismic",jitter =0.3,size = 10,marker="^",alpha=0.4)
plt.legend(loc="upper left")
plt.show()
```

```python
# plot only one value [if i want to plot verticle and horizontally we have to change only x & y]
sns.stripplot(x="total_bill",data=var)
plt.show()
```

# BOX PLOT IN SEABORN

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
var3 = sns.load_dataset("tips")
var3
```

```python
sns.set_style("whitegrid")
sns.boxplot(x="day",y="total_bill",data=var3)
plt.show()
```

```python
# hue parameter in box plot
sns.set_style("whitegrid")
sns.boxplot(x="day",y="total_bill",data=var3,hue ='time')
plt.show()
```

```python
# hue parameter in box plot
sns.set_style("whitegrid")
sns.boxplot(x="day",y="total_bill",data=var3,hue ='sex')
plt.show()
```

```python
# differentiate time and sex together with hue parameter
plt.figure(figsize=(10, 7))
sns.set_style("whitegrid")
var3["time_sex"] = var3["time"].astype(str) + "_" + var3["sex"].astype(str)
sns.boxplot(x="day", y="total_bill", data=var3, hue="time_sex")
plt.show()
```

```python
# palette parameter using as color parameter [`palette='light:g'`] for the same effect
sns.set_style("whitegrid")
sns.boxplot(x="day",y="total_bill",data=var3,hue ='sex',palette='light:g')
plt.show()
```

```python
# palette parameter
sns.set_style("whitegrid")
sns.boxplot(x="day",y="total_bill",data=var3,hue ='sex',palette='viridis')
plt.show()
```

```python
# order parameter
sns.set_style("whitegrid")
sns.boxplot(x="day",y="total_bill",data=var3,hue ='sex',palette='light:b',order=["Sun","Thur","Fri","Sat"])
plt.show()
```

```python
# if you observe the boxplot you can see the mean of the values, now we will plot the mean [showmean parameter]
sns.set_style("whitegrid")
sns.boxplot(x="day",y="total_bill",data=var3,hue ='sex',palette='light:b',order=["Sun","Thur","Fri","Sat"],showmeans=True)
plt.show()
```

```python
# if you want to change something in mean parameter you can make one dictionary meanprops={} and you can do whatever you wnat
sns.set_style("whitegrid")
sns.boxplot(x="day",y="total_bill",data=var3,hue ='sex',palette='light:b',showmeans=True,
            meanprops={"marker":"^","markeredgecolor":"red"})
plt.show()
```

```python
# convert horizontal and vertical [just change x & y parameter]
sns.set_style("whitegrid")
sns.boxplot(y="day",x="total_bill",data=var3,hue ='sex',palette='light:b')
plt.show()
```

```python
# convert horizontal and vertical [if we want to change with orient prameter then both x & y values should be in numerical value ]
plt.figure(figsize=(10, 20))
sns.set_style("whitegrid")
sns.boxplot(y="tip",x="total_bill",data=var3,hue ='sex',palette='light:b', orient="h")
plt.show()
```

# STYLING PLOT IN. SEABORN

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
var6 = sns.load_dataset("tips")
var6
```

# MULTIPLE PLOTS (FACET-GRID) IN SEABORN

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
df = sns.load_dataset('tips')
df
```

```python
fg = sns.FacetGrid(df, col="time", hue="sex")
fg.map(sns.scatterplot, "total_bill", "tip").add_legend()
plt.show()
```

```python
fg = sns.FacetGrid(df, col="day", hue="sex")
fg.map(plt.bar, "total_bill", "tip").add_legend()
plt.show()
```

```python
# edgecolor, palette parameter
fg = sns.FacetGrid(df, col="time", hue="sex",palette="viridis")
fg.map(plt.bar, "total_bill", "tip",edgecolor="g").add_legend()
plt.show()
```
